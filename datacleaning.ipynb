```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
data=pd.read_csv("Bengaluru_House_Data.csv")

data1=data.drop(["area_type","availability","society","balcony"],axis='columns')
data1.head()
```

```python
data1.isnull().sum()
data2=data1.dropna()
data2.isnull().sum()
```

```python
data2["bhk"]=data2["size"].apply(lambda x:int(x.split(" ")[0]))
data2.head()
```

```python
data2.total_sqft.unique()
data3=data2.copy()
```

```python
def is_float(x):
  ele=x.split("-")
  if len(ele)==2:
    return (float(ele[0])+float(ele[1]))/2
  try:
    return float(x)
  except:
    return None
data3["total_sqft"]=data3["total_sqft"].apply(is_float)
data3.head()
```

```python
data3.loc[30]
```

data reductiom

```python
data4=data3.copy()
data4["price_per_unit_sqft"]=data4['price']*100000/data4['total_sqft']
data4.head()
```

```python
data4.location.unique()
data_locations=data4.groupby("location")["location"].agg('count')
data_locations.unique()
len(data_locations[data_locations<=10])
location_less_then_10=data_locations[data_locations<=10]
data5=data4.copy()
data5['location']=data5.location.apply(lambda x : 'other' if x in location_less_then_10 else x )
data5[data5['location']=="other"]
```
