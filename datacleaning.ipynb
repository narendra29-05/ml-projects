```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
data=pd.read_csv("Bengaluru_House_Data.csv")

data1=data.drop(["area_type","availability","society","balcony"],axis='columns')
data1.head()
```

```python
data1.isnull().sum()
data2=data1.dropna()
data2.isnull().sum()
```

```python
data2["bhk"]=data2["size"].apply(lambda x:int(x.split(" ")[0]))
data2.head()
```

```python
data2.total_sqft.unique()
data3=data2.copy()
```

```python
def is_float(x):
  ele=x.split("-")
  if len(ele)==2:
    return (float(ele[0])+float(ele[1]))/2
  try:
    return float(x)
  except:
    return None
data3["total_sqft"]=data3["total_sqft"].apply(is_float)
data3.head()
```
